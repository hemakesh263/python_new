#Create a function which accepts two inputs from the user and compute nCr

```python
def nCr(n,r):
  return(fact(n)/(fact(r)
                *fact(n-r)))
n=int(input(""))
r=int(input(""))
def fact(n):
  res=1
  for i in range(2,n+1):
    res=res*i
  return res
print(int(nCr(n,r)))
```

#Recursive function to compute GCD of 2 numbers

```python
def GCD(a,b):
  if a==b:
    return a
  elif a<b:
    return GCD(b,a)
  else:
    return GCD(b,a-b)

a=int(input("Enter a number:"))
b=int(input("Enter a number:"))
print(GCD(a,b))
```

#Recursive function to find product of two numbers

```python
def prod(a,b):
  if a<b:
    return prod(b,a)
  elif b!=0:
    return (a+prod(a,b-1))
  else:
    return 0
a=int(input("Enter a number:"))
b=int(input("Enter a number:"))
print("The product of a and b is:")
print(prod(a,b))
```

#Recursive function to generate Fibonacci series


```python
def fibo(v):
  if v<=1:
    return v
  else:
    return (fibo(v-1)+fibo(v-2))
v=int(input("Enter a number:"))
print("The fibonacci sequence of entered number is:")
for i in range(v):
  print(i)
print(fibo(v))
```

#Python function that accepts a string and calculates the number of uppercase letters and lowercase letters.

```python
hemakesh="Sky is Blue, my Love is True"
upper=0
lower=0
for i in hemakesh:
  if i.isupper():
    upper+=1
  elif i.islower():
    lower+=1
  else:
    pass
print("uppercase alphabets are:", upper)
print("lowercase alphabets are:", lower)
```
